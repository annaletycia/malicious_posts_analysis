```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('latest_posts_ha_ptbr_prob_relevancia_prev_bin.xlsx')

display(df)
```

```python
df2 = pd.read_excel('latest_posts_deepanswers_ptbr_prob_relevancia_prev_bin.xlsx')

display(df2)
```

# Tratamento de Dados

```python
colunas_selecionadas = ['id', 'full_text', 'probabilidade', 'Relev√¢ncia']

df_filtrado = df[colunas_selecionadas]

display(df_filtrado)
```

```python
df_filtrado2 = df2[colunas_selecionadas]

display(df_filtrado2)
```

```python
linhas_filtradas = df_filtrado[df_filtrado['probabilidade'] >= 0.7]

display(linhas_filtradas)
```

```python
linhas_filtradas2 = df_filtrado2[df_filtrado2['probabilidade'] >= 0.7]

display(linhas_filtradas2)
```

```python
df_unificado = pd.concat([linhas_filtradas, linhas_filtradas2], ignore_index=True)

display(df_unificado)
```

# Utilizando K-Means 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
texts = df_unificado['full_text'].values

tfidf_vectorizer = TfidfVectorizer(max_features=1000) 

tfidf_matrix = tfidf_vectorizer.fit_transform(texts) 
```

```python
num_clusters = 3

kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)

kmeans.fit(tfidf_matrix)
```

```python
df_unificado = df_unificado.copy()

df_unificado.loc[:, 'cluster'] = kmeans.labels_

print(df_unificado.groupby('cluster').size())
```

# Exibindo as palavras mais representativas de cada cluster

```python
centroids = kmeans.cluster_centers_ # matriz onde cada linha representa o centro de um cluster
 
termos = tfidf_vectorizer.get_feature_names_out()
 
num_top_words = 10
 
for i in range(num_clusters):
    print(f"Cluster {i}:")
    # Obtendo os √≠ndices das palavras com maiores valores no centroide do cluster
    top_indices = centroids[i].argsort()[-num_top_words:][::-1]
    top_termos = [termos[index] for index in top_indices]
    print(" ".join(top_termos))
    print("\n")
```

# Visualizando Clusters em um Gr√°fico de Dispers√£o com PCA

```python
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA 
```

```python
pca = PCA(n_components=2, random_state=42)

pca_result = pca.fit_transform(tfidf_matrix.toarray())

df_pca = pd.DataFrame(data=pca_result, columns=['PCA1', 'PCA2'])

df_pca['cluster'] = df_unificado['cluster']

plt.figure(figsize=(10, 7))

for cluster in range(num_clusters):
    cluster_data = df_pca[df_pca['cluster'] == cluster]
    plt.scatter(cluster_data['PCA1'], cluster_data['PCA2'], label=f'Cluster {cluster}', s=50)

plt.title('Clusters visualizados com PCA')
plt.xlabel('PCA Componente 1')
plt.ylabel('PCA Componente 2')
plt.legend()
plt.show()
 
```

#### O PCA √© utilizado para transformar a matriz TF-IDF em duas dimens√µes (PCA1 e PCA2), o que facilita a visualiza√ß√£o de como os documentos se agrupam em um plano 2D.

#### Utiliza-se matplotlib para criar um gr√°fico de dispers√£o onde cada ponto representa um documento, colorido de acordo com o cluster a que pertence.

# Utilizando KNN

```python
from sklearn.neighbors import NearestNeighbors
```

```python
n_neighbors = 5  # ajuste conforme necess√°rio

knn = NearestNeighbors(n_neighbors=n_neighbors, metric='cosine')
knn.fit(tfidf_matrix)

distances, indices = knn.kneighbors(tfidf_matrix)

df_unificado['vizinhos'] = indices.tolist()
df_unificado['distancias'] = distances.tolist()

# Exibindo o DataFrame com os vizinhos
display(df_unificado[['full_text', 'vizinhos', 'distancias']])
```

# Visualiza√ß√£o usando PCA

```python
# Aplicar PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(tfidf_matrix.toarray())

# Criar um DataFrame com os resultados do PCA
df_unificado['pca_x'] = pca_result[:, 0]
df_unificado['pca_y'] = pca_result[:, 1]

# Visualizar os agrupamentos com cores
plt.figure(figsize=(10, 7))
scatter = plt.scatter(df_unificado['pca_x'], df_unificado['pca_y'], 
                      c=df_unificado['cluster'], cmap='viridis', alpha=0.5)

# Adicionar legenda
plt.colorbar(scatter, label='Cluster')
plt.title('Visualiza√ß√£o dos Agrupamentos usando PCA')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.grid()
plt.show()
```

## Utilizando DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=1.2, min_samples=5)  # Experimente esses valores
clusters = dbscan.fit_predict(tfidf_matrix)

df_unificado['dbscan_cluster'] = clusters
```

## Visualiza√ß√£o com PCA

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(tfidf_matrix.toarray())
df_unificado['pca_x'] = pca_result[:, 0]
df_unificado['pca_y'] = pca_result[:, 1]

plt.figure(figsize=(10, 7))
unique_clusters = np.unique(clusters)
colors = plt.colormaps['viridis'](np.linspace(0, 1, len(unique_clusters)))

for i, cluster in enumerate(unique_clusters):
    if cluster == -1:  # -1 √© considerado ru√≠do
        plt.scatter(df_unificado[df_unificado['dbscan_cluster'] == cluster]['pca_x'],
                    df_unificado[df_unificado['dbscan_cluster'] == cluster]['pca_y'],
                    color='red', label='Ru√≠do', alpha=0.5)
    else:
        plt.scatter(df_unificado[df_unificado['dbscan_cluster'] == cluster]['pca_x'],
                    df_unificado[df_unificado['dbscan_cluster'] == cluster]['pca_y'],
                    color=colors[i], label=f'Cluster {cluster}', alpha=0.5)

plt.title('Visualiza√ß√£o dos Agrupamentos usando DBSCAN')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.grid()
plt.show()
```

# Aplicando LDA em cada cluster

```python
from gensim import corpora
from gensim.models import LdaModel
import nltk
from nltk.corpus import stopwords
```

```python
# Carregar stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('portuguese'))
```

```python
for cluster_id in range(num_clusters):
    print(f"T√≥picos do Cluster {cluster_id}:")
    
    cluster_texts = df_unificado[df_unificado['cluster'] == cluster_id]['full_text'].tolist()
    
    texts_processed = [
        [word for word in text.lower().split() if word not in stop_words]
        for text in cluster_texts
    ]
    
    dictionary = corpora.Dictionary(texts_processed)
    corpus = [dictionary.doc2bow(text) for text in texts_processed]
    
    lda_model = LdaModel(corpus, num_topics=5, id2word=dictionary, passes=10)
    
    topics = lda_model.print_topics(num_words=5)
    for topic in topics:
        print(topic)
    
    print("\n")
```
